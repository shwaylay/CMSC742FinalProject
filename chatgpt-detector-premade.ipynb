```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import numpy as np

tokenizer = AutoTokenizer.from_pretrained("Hello-SimpleAI/chatgpt-detector-roberta")

model = AutoModelForSequenceClassification.from_pretrained("Hello-SimpleAI/chatgpt-detector-roberta")
```

```python
inputs = tokenizer("Hello, my dog is cute", return_tensors="tf")
```

```python
text = "There are many different best seller lists that are published by various organizations, and the New York Times is just one of them. The New York Times best seller list is a weekly list that ranks the best-selling books in the United States based on sales data from a number of different retailers. The list is published in the New York Times newspaper and is widely considered to be one of the most influential best seller lists in the book industry. \nIt's important to note that the New York Times best seller list is not the only best seller list out there, and there are many other lists that rank the top-selling books in different categories or in different countries. So it's possible that a book could be a best seller on one list but not on another. \nAdditionally, the term \"best seller\" is often used more broadly to refer to any book that is selling well, regardless of whether it is on a specific best seller list or not. So it's possible that you may hear about a book being a \"best seller\" even if it is not specifically ranked as a number one best seller on the New York Times list or any other list."
encoded_input = tokenizer(text, return_tensors='pt')
output = model(**encoded_input)
```

```python
prediction_logits = output.logits
```

```python
prediction_logits
```

```python
pred = prediction_logits.softmax(dim=-1).tolist()[0]
```

```python
pred = prediction_logits.softmax(dim=-1).tolist()[0]
labels = ["human", "chatgpt"]
labels[np.argmax(pred)]
```

# Reddit Data


```python
results = []
```

```python
import os

base = "/content/drive/MyDrive/CMSC742 Project/Datasets/reddit/train/GPT"

for file in os.listdir(base):
  filename = os.fsdecode(file)
  file_path = os.path.join(base, filename)

  with open(file_path) as f:
    text = f.readlines()[2]
    encoded_input = tokenizer(text, return_tensors='pt')

    output = model(**encoded_input)

    prediction_logits = output.logits
    pred = prediction_logits.softmax(dim=-1).tolist()[0]
    labels = ["human", "chatgpt"]

    results.append([text, "chatgpt", labels[np.argmax(pred)]])
```

```python
base = "/content/drive/MyDrive/CMSC742 Project/Datasets/reddit/test/GPT"

for file in os.listdir(base):
  filename = os.fsdecode(file)
  file_path = os.path.join(base, filename)

  with open(file_path) as f:
    text = f.readlines()[2]
    encoded_input = tokenizer(text, return_tensors='pt')

    output = model(**encoded_input)

    prediction_logits = output.logits
    pred = prediction_logits.softmax(dim=-1).tolist()[0]
    labels = ["human", "chatgpt"]

    results.append([text, "chatgpt", labels[np.argmax(pred)]])
```

```python
base = "/content/drive/MyDrive/CMSC742 Project/Datasets/reddit/train/Real"

for file in os.listdir(base):
  filename = os.fsdecode(file)
  file_path = os.path.join(base, filename)

  with open(file_path) as f:
    text = "".join(f.readlines())
    encoded_input = tokenizer(text, return_tensors='pt')

    output = model(**encoded_input)

    prediction_logits = output.logits
    pred = prediction_logits.softmax(dim=-1).tolist()[0]
    labels = ["human", "chatgpt"]

    results.append([text, "human", labels[np.argmax(pred)]])
```

```python
base = "/content/drive/MyDrive/CMSC742 Project/Datasets/reddit/test/Real"

for file in os.listdir(base):
  filename = os.fsdecode(file)
  file_path = os.path.join(base, filename)

  with open(file_path) as f:
    text = "".join(f.readlines())
    encoded_input = tokenizer(text, return_tensors='pt')

    output = model(**encoded_input)

    prediction_logits = output.logits
    pred = prediction_logits.softmax(dim=-1).tolist()[0]
    labels = ["human", "chatgpt"]

    results.append([text, "human", labels[np.argmax(pred)]])
```

```python
import pandas as pd
reddit = pd.DataFrame(results, columns = ["text", "correct", "predicted"])
```

```python
reddit_correct = len(reddit[reddit["correct"] == reddit["predicted"]])
reddit_incorrect = len(reddit[reddit["correct"] != reddit["predicted"]])
reddit_accuracy = reddit_correct/(reddit_incorrect+reddit_correct)

print ("Correct: " + str(reddit_correct))
print ("Incorrect: " + str(reddit_incorrect))
print ("Accuracy: " + str(round(reddit_accuracy*100,2))+"%")
```
